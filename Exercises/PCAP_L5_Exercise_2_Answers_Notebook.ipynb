## Part 1: Build an `Ingredient` Class with Validation and Default Units

Context: Weâ€™ll start with an `Ingredient` class to represent a single ingredient in the kitchen.

Each ingredient needs:
- `name`     â†’ `"Flour"`, `"Sugar"`, `"Olive oil"`, etc.
- `quantity` â†’ a positive number (e.g. `500`)
- `unit`     â†’ string like `"g"`, `"ml"`, `"pcs"` (default `"g"`)


Your Tasks:
- [ ] 1.1 - Declare an `Ingredient` class.
- [ ] 1.2 - Implement `__init__(self, name, quantity, unit="g")`.
- [ ] 1.3 - If `quantity <= 0`, raise `ValueError("Quantity must be positive")`.
- [ ] 1.4 - On success, store `name`, `quantity`, and `unit` on `self`.

```python
# === Part 1: Your code here ===

# 1.1)
class Ingredient:
    # 1.2)
    def __init__(self, name, quantity, unit="g"):
        """
        Initialize an ingredient with a name, quantity, and unit.

        Quantity must be positive.
        """
        # 1.3)
        if quantity <= 0:
            raise ValueError("Quantity must be positive")
        # 1.4)
        self.name = name
        self.quantity = quantity
        self.unit = unit
```

## Part 2: `Recipe` with default difficulty and ingredient list

Context: Next, we'll create a `Recipe` class that groups ingredients together.

Each recipe needs:
- `name`       â†’ the recipe name, like `"Pasta Carbonara"`
- `difficulty` â†’ `"Easy"`, `"Medium"`, `"Hard"` (default `"Medium"`)
- `ingredients` â†’ a list to hold `Ingredient` objects

Your Tasks:
- [ ] 2.1 - Declare a `Recipe` class.
- [ ] 2.2 - Implement `__init__` function with input parametgers `self`, `name`, and `difficulty` (where `difficulty` has a default of `"Medium"`):
  - store `name` and `difficulty`
  - initialize `self.ingredients` with an empty list
- [ ] 2.3 - Add a method `add_ingredient(self, ingredient)` that:
  - appends the `ingredient` to `self.ingredients`

```python
# === Part 2: Your code here ===

# 2.1)
class Recipe:
    # 2.2)
    def __init__(self, name, difficulty="Medium"):
        self.name = name
        self.difficulty = difficulty
        self.ingredients = []

    # 2.3)
    def add_ingredient(self, ingredient):
        self.ingredients.append(ingredient)
```

## Part 3: `Dish` with Validation, Class Attribute, and `@classmethod` Factory


Context: Finally, we'll define a `Dish` that goes on the menu.

Each `Dish` needs:
- `name`   â†’ name of the dish (e.g. `"Margherita Pizza"`)
- `chef`   â†’ name of the chef responsible
- `price`  â†’ positive number

We'll also:
- Use a class attribute `tax_rate` (e.g. `0.1` for 10%).
- Provide a `final_price()` instance method that applies tax.
- Provide a `@classmethod create_daily_special(cls, base_name, chef)`  
  to quickly build a â€œDaily Specialâ€ with a fixed base price.


Your Tasks:
- [ ] 3.1 - Declare a `Dish` class with a class attribute `tax_rate = 0.1`.
- [ ] 3.2 - Implement `__init__(self, name, chef, price)`:
  - if `price <= 0`, raise `ValueError("Price must be positive")`
  - otherwise store `name`, `chef`, and `price`
- [ ] 3.3 - Implement the `final_price(self)` calculation:
  - return the value of `self.price` x `(1 + self.tax_rate)`
- [ ] 3.4 - Implement `@classmethod create_daily_special(cls, base_name, chef)`:
  - build `special_name = f"Daily Special: {base_name}"`
  - create and return `cls(special_name, chef, price=20.0)`

```python
# === Part 3: Your code here ===

# 3.1)
class Dish:
    tax_rate = 0.1

    # 3.2)
    def __init__(self, name, chef, price):
        """
        Create a new dish with a positive price.
        """
        if price <= 0:
            raise ValueError("Price must be positive")
        self.name = name
        self.chef = chef
        self.price = price

    # 3.3)
    def final_price(self):
        """
        Return the price including tax.
        """
        return self.price * (1 + self.tax_rate)

    # 3.4)
    @classmethod
    def create_daily_special(cls, base_name, chef):
        """
        Create a daily special with a fixed base price.
        """
        special_name = f"Daily Special: {base_name}"
        return cls(special_name, chef, price=20.0)
```

## Part 5: Debrief (short answers)

### Q1 of 2: Why is it helpful to do validation (like checking price or quantity) inside of `__init__()`?

It's because validation inside of `__init__()` makes sure that objects are never created in a broken state.

More specifically:
- If the data is invalid, the object is not fully constructed and a clear error is raised.
- Any code that does have an instance can trust its attributes (`price`, `quantity`, etc.) to be valid.

Because of this, it leads to fewer hidden bugs and makes your objects safer to use throughout the program.

_______

### Q2 of 2: How is `Dish.create_daily_special(...)` different from calling `Dish(...)` directly?

`Dish(...)` is the normal constructor, meaning you must supply all details yourself (`name`, `chef`, and `price`) every single time.

`Dish.create_daily_special(...)` on the other hand is an alternative constructor, meaning:

- It's a `@classmethod`, so it receives `cls` instead of `self`.
- It chooses a standard pattern for you (name prefix + fixed base price).
- It still returns a fully initialized `Dish` instance, but with some decisions pre-made.

This is handy method in Python to know when you want multiple, convenient ways to build valid objects.
